```python
import re
def remove_emoji(text: str):
    regex_pattern = re.compile(pattern="["
u"\U0001F600-\U0001F64F" # emoticons
u"\U0001F300-\U0001F5FF" # symbols & pictographs
u"\U0001F680-\U0001F6FF" # transport & map symbols
u"\U0001F1E0-\U0001F1FF" # flags (iOS)
u"\U00002500-\U00002BEF" # chinese char
u"\U00002702-\U000027B0"
u"\U00002702-\U000027B0"
u"\U000024C2-\U0001F251"
u"\U0001f926-\U0001f937"
u"\U00010000-\U0010ffff"
u"\u2640-\u2642"
u"\u2600-\u2B55"
u"\u200d"
u"\u23cf"
u"\u23e9"
u"\u231a"
u"\ufe0f" # dingbats
u"\u3030"
"]+", flags=re.UNICODE)
    
    return regex_pattern.sub(r'', text)
```

```python
app_info['descriptionHTML'] = app_info['descriptionHTML'].apply(remove_emoji)
```

```python
import re

def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
```

## Bigrams, Trigrams,...

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

# use English stopwords
stops = set(stopwords.words("english"))
```

```python
# convert the relevant column to lowercase since nltk works with lists, not dataframes
app_info['descriptionHTML'] = app_info.descriptionHTML.str.lower()

# create a list of the reviews from the 'Reviews' column
reviews = app_info['descriptionHTML'].tolist()
```

```python
reviews[0:2]
```

```python
# tokenise the words
from nltk.tokenize import word_tokenize

content_tokens = []
for word in reviews:
    content_tokens.append(word_tokenize(word))
```

```python
content_tokens[0]
```

```python
# Using the stopwords above, create a new list with the stopwords removed from word_tokens
filtered_content = []
for word in content_tokens:
    filtered_content.append([w for w in word if not w in stops])
    
# adding all these new tokens to a new column in df
app_info['filtered_content'] = filtered_content
```

```python
app_info.head()
```

```python
# Finally, put all the tokens back together as a text:
import string
joined_tokens = []
for word in filtered_content:
    z = ",".join(word) # join all the stemmed words back together on the commas
    z = z.replace(",", " ") # replace all the commas with spaces
    # remove the punctuation from the text:
    joined_tokens.append(z.translate(str.maketrans('', '')))
    
# add this new and reformed text to the data frame
app_info['Joined token reviews'] = joined_tokens
app_info['Joined token reviews'] = app_info['Joined token reviews'].map(lambda x: re.sub('[(),\.!?#&*:;]', '', x))
```

## WordCloud

```python
from wordcloud import WordCloud, STOPWORDS 
```

```python
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in app_info['Joined token reviews']: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 14).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()  
```

### Stemming

```python
# Now on to stemming: most common used is PorterStemmer (manag = 'manager')
from nltk.stem import PorterStemmer
ps = PorterStemmer()

stemmed_descriptions = []
for word in filtered_content:
    stemmed_descriptions.append([ps.stem(w) for w in word])
    
# add these to the dataframe, df
app_info['half_stemmed'] = stemmed_descriptions
```

```python
stemmed_descriptions[0]
```

```python
# Finally, put all the tokens back together as a text:
import string
joined_stem = []
for word in stemmed_descriptions:
    z = ",".join(word) # join all the stemmed words back together on the commas
    z = z.replace(",", " ") # replace all the commas with spaces
    # remove the punctuation from the text:
    joined_stem.append(z.translate(str.maketrans('', '')))
    
# add this new and reformed text to the data frame
app_info['Stemmed_descriptions'] = joined_stem
app_info['Stemmed_descriptions'] = app_info['Stemmed_descriptions'].map(lambda x: re.sub('[(),\.!?#&*:;]', '', x))
```

```python
app_info.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# set vectorizer
vectorizer = TfidfVectorizer()

# array of word counts
document_vec = vectorizer.fit_transform(list(app_info['Joined token reviews']))

# convert to dataframe
df2 = pd.DataFrame(document_vec.toarray(), columns=vectorizer.get_feature_names())

# set a threshold to drop infrequent words
threshold = 0.1

# drop words based on the threshold
df2 = df2.drop(df2.mean()[df2.mean() < threshold].index.values, axis=1) # axis=1 for column mean
```

```python
app_info = app_info.join(df2, how='left')
```

```python
app_info.columns
```

```python
# We use df1 as it contains no links
# Find all the bigrams

from nltk import bigrams

# Find the best 'best' bigrams / trigrams

from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
quadgram_measures = nltk.collocations.QuadgramAssocMeasures()

# create a single string that combines 200 reviews
# words is the list of reviews created earlier
long_string = word_tokenize(",".join(list(app_info['Joined token reviews'])[0:200]))

# Bigrams
finder = BigramCollocationFinder.from_words(long_string)
# only bigrams that appear >= 2x
finder.apply_freq_filter(2)
# return the 10 bigrams with the highest likelihood
print("Bigrams of app descriptions\n")
print(finder.nbest(bigram_measures.likelihood_ratio, 30))
```

```python
# Trigrams
finder = TrigramCollocationFinder.from_words(long_string)
# only bigrams that appear >= 2x
finder.apply_freq_filter(2)
# return the 10 trigrams with the highest likelihood
print("Trigrams of app descriptions\n")
print(finder.nbest(trigram_measures.likelihood_ratio, 20))
```

```python
# Quadgrams
quadgram_measures = nltk.collocations.QuadgramAssocMeasures()

finder = QuadgramCollocationFinder.from_words(long_string)
# only bigrams that appear >= 2x
finder.apply_freq_filter(2)
# return the 10 trigrams with the highest likelihood
print("Quads of app descriptions\n")
print(finder.nbest(quadgram_measures.likelihood_ratio, 10))
```
